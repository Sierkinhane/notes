## generate heatmap

```python
# gaussian type
def generate_heatmap(background, pt, sigma, label_type='Gaussian'):
    # check that any part of the gassian is in-bounds
    tmp_size = sigma * 3
    # pt = [x,y]
    upper_left = [int(pt[0] - tmp_size), int(pt[1] - tmp_size)]
    bottom_right = [int(pt[0] + tmp_size + 1), int(pt[1] + tmp_size + 1)]
    if (upper_left[0] >= background.shape[1] or upper_left[1] >= background.shape[0] or bottom_right[0] < 0 or bottom_right[1] < 0):
        return img
    """
    generate gaussian heatmap
        heatmapheatmap
    """
    size = 2 * tmp_size + 1
    x = np.arange(0, size, 1, np.float32)
    y = x[:, np.newaxis]
    x0 = y0 = size // 2
    # the gaussian is not normalized, we want the center value to equal to 1
    if label_type == 'Gaussian':
        g = np.exp(-((x-x0)**2 + (y-y0)**2) / (2 *sigma **2))
    else:
        pass
    # Usable gaussian range
    g_x = max(0, -upper_left[0]), min(bottom_right[0], background.shape[1]) - upper_left[0]
    g_y = max(0, -upper_left[1]), min(bottom_right[1], background.shape[0]) - upper_left[1]
    # Image range
    img_x = max(0, upper_left[0]), min(bottom_right[0], background.shape[1])
    img_y = max(0, upper_left[1]), min(bottom_right[1], background.shape[0])

    img[img_y[0]:img_y[1], img_x[0]:img_x[1]] = g[g_y[0]:g_y[1], g_x[0]:g_x[1]]
    return img

def random_shift(img, max_pixel, ori_peak_coords):
    """
    shift the heatmap
    img = (h, w), ndarray
    lr:  >0 right; <0 left [-mp, mp]
    ud:  >0 up;    <0 down [-mp, mp]
    """
    if len(img.shape) == 3:
        img = np.squeeze(img)
    lr = np.random.randint(-max_pixel, max_pixel) # 5
    ud = np.random.randint(-max_pixel, max_pixel) # 6
    # lr=20
    # ud=20
    print(ud, lr)
    h, w = img.shape[:2]
    padding_img = np.zeros((h+np.abs(2*ud), w+np.abs(2*lr)))
    padding_img[np.abs(ud):np.abs(ud)+64,np.abs(lr):np.abs(lr)+64] = img
    if ud>=0 and lr>=0:
        shift = padding_img[0:64,0:64]
        peak_coords = [ori_peak_coords[0] + lr, ori_peak_coords[1] + ud]
    elif ud>=0 and lr<0:
        shift = padding_img[0:64,2*np.abs(lr):]
        peak_coords = [ori_peak_coords[0] - np.abs(lr), ori_peak_coords[1] + np.abs(ud)]
    elif ud<0 and lr<0:
        shift = padding_img[2*np.abs(ud):,2*np.abs(lr):]
        peak_coords = [ori_peak_coords[0] - np.abs(lr), ori_peak_coords[1] - np.abs(ud)]
    else: # ud <0 lr >0
        shift = padding_img[2*np.abs(ud):,0:64]
        peak_coords = [ori_peak_coords[0] + np.abs(lr), ori_peak_coords[1] - np.abs(ud)]
    
    return shift, np.array(peak_coords)
```

## heatmap vis


```python
# 2. show input and target
def data_visulization(score_map, pt):
    """
    score map must shape like (N, H, W) and its type should be ndarray.
    pt must shape like (N, 2) and its type should be ndarray.
    N: num of heatmap
    """
#     if instance(pt):
#     print(instance(pt))
    if len(score_map.shape) == 4:
        score_map = np.squeeze(score_map)
    elif len(score_map.shape) == 2:
        score_map = np.expand_dims(score_map, axis=0)
        
    if len(pt.shape) == 1:
        pt = np.expand_dims(pt, axis=0)
#     print(pt.shape)
    for i in range(0, score_map.shape[0]):
        m = score_map[i,:,:]
        colormap = cv2.applyColorMap(m, cv2.COLORMAP_HSV)
#         plt.axis('off')
#         plt.imshow(img)
        plt.imshow(colormap, alpha=0.5)
        plt.plot(pt[i,0], pt[i,1], 'r.', ms=5)
#         plt.savefig('../../Heatmap-finetune-speed-up-network/heatmaps/pred_colormap_' + str(i) + '.jpg')
        plt.show()
        plt.close()
```

```python
img = np.zeros((64,64))
pt = np.array([10,32]) # x,y
sigma = 1.5
img = generate_heatmap(img, pt, sigma) 
data_visulization(img, pt)
# img = (generate_target(img, pt, sigma)*255.).astype(np.uint8)
# img = ndi.grey_dilation(img, size=(3,3))
img, peak_coords = random_shift(img, 5, pt)
data_visulization(img, peak_coords)
print("peak_coords is {0}, and its value is {1}.".format(peak_coords, img[peak_coords[0], peak_coords[1]]))
```

```python
def show_keypoint(img, pt):
    """
    img(ndarray): RGB or GRAY, shape like (h, w, c)
    pt(ndarray): shape like (N, 2)
    """
    plt.imshow(img, cmap='gray')
    plt.plot(pt[:, 0], pt[:, 1], 'g.', ms=7)
    plt.show()
    plt.close()
```

```python
def normalize(img):
    """
    normalize
    """
    mean = np.array([0.485, 0.456, 0.406], dtype=np.float32)
    std = np.array([0.229, 0.224, 0.225], dtype=np.float32)
    img = img.astype(np.float32)
    img = (img/255.0 - mean) / std
    img = img.transpose([2, 0, 1])
    img = np.expand_dims(img, axis=0)
    
    return img
```

```python
from network.HRNet import get_face_alignment_net
from network.FC import Fc
from network.unet.unet_model import UNet_ultras
from network.config import config, update_config
import yaml
import time
import torch

from yacs.config import CfgNode as CN

with open('network/hrnet.yaml', 'r') as f:
    config = CN.load_cfg(f.read()) 
    config.freeze()
#     print(config['MODEL'])

HRNet = get_face_alignment_net(config).cuda().eval()
fc = Fc().cuda().eval()
unet = UNet_ultras(1, 1).cuda().eval()

HRNet.load_state_dict(torch.load('checkpoints/HR18-WFLW.pth'))
fc.load_state_dict(torch.load('checkpoints/model_best_fc_850_[hr pose].pth'))
unet.load_state_dict(torch.load('checkpoints/model_best_unet_us_759_[hrnet pose].pth')['state_dict'])

# img = cv2.imread('images/61_Street_Battle_streetfight_61_521_86.jpg')
img = cv2.imread('images/leslie2.jpg')
img = cv2.resize(img, (256, 256))
face = img.copy()
img = normalize(img)

img = torch.Tensor(img).cuda()
s = time.time()
heatmap = HRNet(img).permute(1, 0, 2, 3)
heatmap = unet(heatmap)
coords = fc(heatmap).detach()
e = time.time()
coords = coords.cpu().numpy()
show_keypoint(face, coords*4)
heatmap = heatmap.detach()
heatmap = np.squeeze(heatmap.cpu().numpy())
data_visulization(heatmap, coords-1)
print('elapsed time: {:6f}s'.format(e-s))
```
