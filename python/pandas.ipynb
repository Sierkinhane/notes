## 将某个特征数量化

```python
df['BusinessTravel'].value_counts()
d = {'Travel_Rarely':0, 'Travel_Frequently':1, 'Non-Travel':2}
df['BusinessTravel'] = df['BusinessTravel'].map(d)
df['BusinessTravel'][:10]
```

## 将非数字的特征数量化

```python
df.describe(include='object').columns
```

```python
df.info
```

```python
# df = pd.read_csv("./data/pfm_train.csv")
# 特征名称
for feature in df.describe(include='object').columns:
    # 每个特征里面的变量
    var = df[feature].value_counts().index.tolist()
    d = {}
    for i, v in enumerate(var):
        d[v] = i
    df[feature] = df[feature].map(d)
df.info()
df.to_csv('./data/digitalized_train.csv')
```
